```python
#proyek latihan hal 57 no.8
#input
s=float(input('Strawberry(kg):'))
j=float(input('Jeruk(kg):'))
an=float(input('anggur(kg):'))
ap=float(input('apel(kg)'))
se=float(input('semangka(kg):'))
#harga
S=s*30000
J=j*15000
AN=an*42000
AP=ap*38000
SE=se*21000
#Total harga
total=S+J+AN+AP+SE
print('Pendapatan :Rp',total)
#profit
PS=S-(25000*s)
PJ=J-(12500*j)
PAN=AN-(39000*an)
PAP=AP-(35000*ap)
PSE=SE-(19200*se)
P=PS+PJ+PAN+PAP+PSE
print('profit:Rp',P)
```

```python
#proyek latihan hal 58 no.10
x=input('masukan no.komputer:')
#waktu mulai akses
print('waktu mulai akses')
jm=float(input('jam:'))
mm=float(input('menit:'))
dm=float(input('detik:'))
#waktu selesai akses
print('waktu selesai akses')
js=float(input('jam:'))
ms=float(input('menit'))
ds=float(input('detik'))
#total waktu akses
tdm=jm*3600+mm*60+dm
tds=js*3600+ms*60+ds
selisih=tds-tdm
tj=selisih//3600
s1=selisih%3600
tm=s1//60
td=s1%60
print('total waktu akses:',tj,'jam',tm,'menit',td,'detik')
#total biaya akses 
biaya=(selisih/(30*60))*2250
print('total biaya akses no komputer',x,'Rp :',biaya)
```
